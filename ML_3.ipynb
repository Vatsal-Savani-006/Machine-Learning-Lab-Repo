##Study and implement PCA in python.

Load the wine dataset from sklearn library

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
print("Initial Shape of our Dataset", X.shape)
```

Standardization of the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Covariance Matrix

```python
cov_matrix = np.cov(X_scaled.T)  # Transpose because numpy expects features in rows
print("Covariance matrix shape:", cov_matrix.shape)
```

Eigenvalues and Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors shape:", eigenvectors.shape)
```

Sort Eigenvalues and Eigenvectors

```python
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]
```

Calculate Explained Variance Ratio

```python
explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)

print("\nExplained Variance Ratio:")
for i, var in enumerate(explained_variance_ratio):
    print(f"PC{i+1}: {var:.4f}")
```

Select Top k Components

```python
k = 7
top_k_eigenvectors = sorted_eigenvectors[:, :k]
```

Transform the Data (Projection)

```python
X_pca = X_scaled.dot(top_k_eigenvectors)

print("\nShape after PCA transformation:", X_pca.shape)
print("First 5 rows of transformed data:\n", X_pca[:5])
```

PCA with sklearn

```python
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(X_scaled)
print("\nExplained Variance Ratio:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"PC{i+1}: {var:.4f}")
```
