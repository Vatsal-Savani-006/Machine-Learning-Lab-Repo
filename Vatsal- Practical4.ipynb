#Dataset Preparation

```python
import pandas as pd
import numpy as np
df=pd.read_csv(path + "/SOCR-HeightWeight.csv")
df.head()
```

#Normal Equation

```python
from numpy import sign
def intercept_slop(x,y):
    n=len(x)
    sig_x=np.sum(x)
    sig_y=np.sum(y)
    sign_xy=np.sum(x*y)
    sig_x2=np.sum(x**2)
    m = (n*sign_xy-sig_x*sig_y)/(n*sig_x2-sig_x**2)
    c = (sig_y-m*sig_x)/n
    return m,c

m,c = intercept_slop(df["Height(Inches)"],df["Weight(Pounds)"])
print(m,c)
```

##Prediction

```python
def predict(x,m,c):
    y_pred=m*x+c
    y_pred=pd.Series(y_pred,name="Weight(Pounds)")
    return y_pred
y_pred=predict(df["Height(Inches)"],m,c)
y_pred.head()
```

##Mean Squared Error

```python
from sklearn.metrics import r2_score, mean_squared_error
y_org = df['Weight(Pounds)']
print('Mean Squared Error:', mean_squared_error(y_pred, y_org))
```
